```python
def scrape(inicio=56591,fim=56491):
    
    path='C:/Users/Felipe/Documents/GitHub/IntroDS-Projeto-Final/Data/'
    browser=webdriver.Firefox()
    url1='https://www.camara.leg.br/internet/escriba/escriba.asp?codSileg={0}'.format(inicio)
    browser.get(url1)
    iframe=browser.find_element_by_xpath('//*[@id="ifrmEscriba"]')
    browser.switch_to.frame(iframe)
    #Sessoes=[]
    Verificar=[]
    
    for i in range(inicio,fim,-1):
        try:
            temp={}
            url='https://www.camara.leg.br/internet/escriba/escriba.asp?codSileg={0}'.format(i)
            browser.get(url)
            iframe=browser.find_element_by_xpath('//*[@id="ifrmEscriba"]')
            browser.switch_to.frame(iframe)
            
            sessao=browser.find_element_by_xpath('/html/body/div/div[1]/div/strong[3]').text
            data=browser.find_element_by_xpath('/html/body/div/div[1]/div/span[1]/span').text
            tipo=browser.find_element_by_xpath('/html/body/div/div[1]/div/strong[4]').text
            
            tudo='//*[@id="contentEncontro"]'
            temp['Sessao']=sessao
            temp['Data']=data
            temp['Tipo']=tipo
            temp['Pagina']=i
            
            conteudo=browser.find_element_by_xpath(tudo).text
            temp['Conteudo']=conteudo
            #Sessoes.append(temp)
            
            file_name=data[-4:]+'_'+sessao
            file=os.path.join(path,file_name+'.txt')
            
            with open(file,'w') as file:
                file.write(json.dumps(temp))
        
            
        except:
            
            try:
                browser.find_element_by_xpath('/html/body/div/h1').text
            except:
                Verificar.append(i)
                
                
    file_name_v='Pendente'+'_'+str(date.today())+'_'+str(datetime.now().strftime('%H:%M')).replace(':','_')
    file_v=os.path.join(path,file_name_v+'.txt')    
    with open(file_v,'w') as file:
                file.write(json.dumps(Verificar))
    return Verificar
```

```python
start=time.time()
v=scrape(50706,fim=47500)
end=time.time()
print(end-start)
```
