```python
import time
from selenium import webdriver
from numpy import NaN
```

```python
def scrape(deputado):

    browser=webdriver.Firefox()


    browser.get('https://www2.camara.leg.br/atividade-legislativa/discursos-e-notas-taquigraficas')
    btn2=browser.find_element_by_xpath('//*[@id="txOrador"]')
    btn2.send_keys(deputado)
    btn=browser.find_element_by_xpath('/html/body/div[2]/div[2]/div[2]/div/div[2]/div[3]/form/input[1]')
    btn.click()
    browser.switch_to.alert.accept()
    
    time.sleep(5)
    
    qtd=browser.find_element_by_xpath('/html/body/div[1]/div[1]/div[2]/div/div/div/div/div[2]/div/span[3]').text
    qtd=qtd.replace('.','')
    nr_doc=int(qtd)
    paginas=int(nr_doc/50)+1
    Discursos=[]
    

    
    for pag in range(1,paginas+1):
            
        if pag==1:
            pass
        elif pag==2:
            proxpag=browser.find_element_by_xpath('/html/body/div[1]/div[1]/div[2]/div/div/div/div/div[2]/div/div[3]/span/a[2]')
            proxpag.click()
        else:
            proxpag=browser.find_element_by_xpath('/html/body/div[1]/div[1]/div[2]/div/div/div/div/div[2]/div/div[3]/span[2]/a[2]')
            proxpag.click()
            
        if pag==paginas:
            td=(nr_doc-50*(pag-1))*2+1
        else:
            td=101
        
        for doc in range(1,td,2):
                
            temp={}
            
            caminho1='/html/body/div[1]/div[1]/div[2]/div/div/div/div/div[2]/div/table/tbody/tr[{0}]/td[1]'.format(doc)
            data=browser.find_element_by_xpath(caminho1).text
            temp['Data']=data
                
            try:
                caminho2='/html/body/div[1]/div[1]/div[2]/div/div/div/div/div[2]/div/table/tbody/tr[{0}]/td[2]'.format(doc)
                sessao=browser.find_element_by_xpath(caminho2).text
                temp['Sessão']=sessao
            except:
                temp['Sessão']=NaN
                    
            try:
                caminho3='/html/body/div[1]/div[1]/div[2]/div/div/div/div/div[2]/div/table/tbody/tr[{0}]/td[3]'.format(doc)
                fase=browser.find_element_by_xpath(caminho3).text
                temp['Fase']=fase
            except:
                temp['Fase']=NaN
                    
            try:
                caminho4='/html/body/div[1]/div[1]/div[2]/div/div/div/div/div[2]/div/table/tbody/tr[{0}]/td[4]/a/span'.format(doc)
                discurso=browser.find_element_by_xpath(caminho4)
                discurso.click()
                conteudo=browser.find_element_by_xpath('//*[@id="content"]').text
                temp['Discurso']=conteudo
                browser.back()
            except:
                temp['Discurso']=NaN
                    
                    
            try:
                caminho5='/html/body/div[1]/div[1]/div[2]/div/div/div/div/div[2]/div/table/tbody/tr[{0}]/td[7]'.format(doc)
                hora=browser.find_element_by_xpath(caminho5).text
                temp['Hora']=hora
            except:
                temp['Hora']=NaN
                
            Discursos.append(temp)
            print(doc)
       
        
        
    
    return Discursos
        
```

```python
start=time.time()
f=scrape('Ciro Nogueira')
end=time.time()
print(end-start)
```
